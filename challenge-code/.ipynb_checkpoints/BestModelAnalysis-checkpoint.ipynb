# Analysing the Best Model

```python
# Imports
import matplotlib.pyplot as plt
import osmnx as ox

import heapq
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import defaultdict
from math import floor, ceil, inf
from dataclasses import asdict

from edinburgh_challenge.constants import police_stations, police_stations_dict
from edinburgh_challenge.utility import generate_early_shift_distributions
from edinburgh_challenge.models import NaiveModel, GreedyModel, EnhancedModel, SimplifiedModelNotBest, SimplifiedModel
from edinburgh_challenge.simulation import *
from edinburgh_challenge.processing import calculate_metric, calculate_simulation_performance
```

```python
source = "./data.xlsx"
data = pd.read_excel(source)
data["Time"] = (data["Day"]-1)*24 + data["Hour"]
data.columns = [x.lower() for x in data.columns]
```

```python
ps = [[f'PS_{i}', p.x, p.y] for i,p in 
      enumerate([police_stations.one, 
                 police_stations.two, 
                 police_stations.three])]
df_ps = pd.DataFrame(ps, columns=["Name","Latitude", "Longitude"])
```

```python
# Running the best model
shift_distribution = {'Early': {'Station_1': 0, 'Station_2': 5, 'Station_3': 10},
  'Day': {'Station_1': 2, 'Station_2': 19, 'Station_3': 4},
  'Night': {'Station_1': 9, 'Station_2': 3, 'Station_3': 28}
}

ps_coords = [ (p.x, p.y) for p in 
                [police_stations.one, 
                 police_stations.two, 
                 police_stations.three]]

simulation = SimulationWithMaxUtilisation(data, ps_coords, shift_distribution, 
                        verbose=0)



greedy_model = GreedyModel(shift_distribution, police_stations_dict)
```

```python
simulation.run(greedy_model)
print(f"Performance: {calculate_simulation_performance(simulation.analyze_simulation_results())}")
```

## Which of the cases did the model miss?

```python
missed_incidents = {
    "Immediate":[],
    "Prompt":[],
    "Standard":[],
}

thresholds = {'Immediate': 1, 'Prompt': 3, 'Standard': 6}
for incident in simulation.cumulative_incidents:
    priority = incident.priority
    time_to_response = incident.response_time - incident.global_time

    if time_to_response > thresholds[priority]:
        missed_incidents[priority].append(incident)
        
    
```

```python
missed_immediate = missed_incidents["Immediate"]
missed_prompt = missed_incidents["Prompt"]
missed_standard = missed_incidents["Standard"]
all_missed = missed_immediate + missed_prompt + missed_standard
all_incidents = simulation.cumulative_incidents

incidents_data = [asdict(incident) for incident in all_missed]
all_incidents_data = [asdict(incident) for incident in all_incidents]
incidents_df = pd.DataFrame(incidents_data)
all_incidents_df = pd.DataFrame(all_incidents)
```

```python
sns.histplot(data=all_incidents_df, x='global_time', bins=15, palette=["black"], label="No. of Incidents")
plt.title('Distribution of Missed Incidents over Time')
plt.xlabel('Global Time (in hours)')
plt.ylabel('Number of Incidents')
plt.legend()
plt.savefig("pictures/distribution_of_incidents.png")
plt.show()
```

## Officer Hours spent distribution

```python
total_officer_hours = {}

for incident in simulation.cumulative_incidents:
    if incident.resolved:
        if incident.resolving_officer not in total_officer_hours.keys():
            total_officer_hours[incident.resolving_officer]  = 0

        time_spent_on_incident = incident.resolution_time - incident.global_time
        total_officer_hours[incident.resolving_officer] += time_spent_on_incident
```

## Resource Allocation Graph

```python
priority_allocations = {'Immediate': [], 'Prompt': [], 'Standard': []}

# Iterate through cumulative incidents and gather data by priority
for incident in simulation.cumulative_incidents:
    priority = incident.priority
    resolution_time = incident.resolution_time
    time_of_allocation = incident.global_time + incident.response_time
    
    # Append data to the corresponding priority list
    priority_allocations[priority].append((incident.global_time, resolution_time, time_of_allocation))

# Create separate resource allocation graphs for each priority
plt.figure(figsize=(12, 6))

for priority, data in priority_allocations.items():
    if data:
        data.sort(key=lambda x: x[0])  # Sort data by time_step
        time_steps, resolution_times, allocation_times = zip(*data)
        plt.plot(time_steps, allocation_times, label=priority)

# Customize the plot
plt.title('Resource Allocation Over Time by Priority')
plt.xlabel('Time Step')
plt.ylabel('Time of Allocation')
plt.legend()
plt.grid(True)
plt.show()
```

## Comparison of different models

```python
def initialise_simulation():
    shift_distribution = {'Early': {'Station_1': 0, 'Station_2': 5, 'Station_3': 10},
      'Day': {'Station_1': 2, 'Station_2': 19, 'Station_3': 4},
      'Night': {'Station_1': 9, 'Station_2': 3, 'Station_3': 28}
    }
    
    ps_coords = [ (p.x, p.y) for p in 
                        [police_stations.one, 
                         police_stations.two, 
                         police_stations.three]]
        
    simulation = SimulationWithMaxUtilisation(data, ps_coords, shift_distribution, 
                                verbose=0)
    return simulation
```

## Greedy Model

```python
shift_distribution = {'Early': {'Station_1': 0, 'Station_2': 5, 'Station_3': 10},
      'Day': {'Station_1': 2, 'Station_2': 19, 'Station_3': 4},
      'Night': {'Station_1': 9, 'Station_2': 3, 'Station_3': 28}
    }
simulation = initialise_simulation()       
greedy_model = GreedyModel(shift_distribution, police_stations_dict)
simulation.run(greedy_model)
print(f"Performance: {calculate_simulation_performance(simulation.analyze_simulation_results())}")
```
