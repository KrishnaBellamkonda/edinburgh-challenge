```python
def convert_dataframe(df):
    """
    Convert a DataFrame with Easting and Northing columns to Latitude and Longitude.
    
    :param df: DataFrame with columns 'X' (Easting) and 'Y' (Northing)
    :return: DataFrame with additional columns 'Latitude' and 'Longitude'
    """
    latitudes = []
    longitudes = []
    
    for index, row in df.iterrows():
        lat, lon = bng_to_latlong(row['X (Easting)'], row['Y (Northing)'])
        latitudes.append(lat)
        longitudes.append(lon)
    
    df['Latitude'] = latitudes
    df['Longitude'] = longitudes
    
    return df
```

### Understanding the simulation

```python
missed_incidents = {
    "Immediate":[],
    "Prompt":[],
    "Standard":[],
}

thresholds = {'Immediate': 1, 'Prompt': 3, 'Standard': 6}
for incident in simulation.cumulative_incidents:
    priority = incident.priority
    time_to_response = incident.response_time - incident.global_time

    if time_to_response > thresholds[priority]:
        missed_incidents[priority].append(incident)
        
    
```

```python
missed_immediate = missed_incidents["Immediate"]
missed_prompt = missed_incidents["Prompt"]
missed_standard = missed_incidents["Standard"]
all_missed = missed_immediate + missed_prompt + missed_standard
all_incidents = simulation.cumulative_incidents

incidents_data = [asdict(incident) for incident in all_missed]
all_incidents_data = [asdict(incident) for incident in all_incidents]
incidents_df = pd.DataFrame(incidents_data)
all_incidents_df = pd.DataFrame(all_incidents)
```
